#Recurrent Neural Network(RNN)

Recurrent Neural Network(RNN) is a type of Neural Network where the output from the previous step is fed as input to the current step. In traditional neural networks, all the inputs and outputs are independent of each other, but in cases when it is required to predict the next word of a sentence, the previous words are required and hence there is a need to remember the previous words. Thus RNN came into existence, which solved this issue with the help of a Hidden Layer. The main and most important feature of RNN is its Hidden state, which remembers some information about a sequence

RNN works very well sequence of data as input.so it is use in spam classifier,time series,weather forcasting

```python
def change(message):
  if message == "not happy":
    return 0
  else:
    return 1
```

```python
# Initialize NLTK resources (you might need to download NLTK resources)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Define a lemmatizer and set of stopwords
lem = WordNetLemmatizer()
st = set(stopwords.words("english"))

wd = []

# Ensure 'data' is defined
if 'data' in globals():
    for i in data['Description']:
        sent = re.sub('[^A-Za-z]', " ", i)
        words = nltk.word_tokenize(sent)
        unique_words = set()
        for w in words:
            if w.lower() not in st:  # Convert to lowercase for stopwords check
                lw = lem.lemmatize(w)
                unique_words.add(lw)
        clean_statements = " ".join(unique_words)
        wd.append(clean_statements)
else:
    print("'data' is not defined. Please define 'data' before running this code.")

# 'wd' now contains the preprocessed descriptions
```

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
len(wd)
```

```python
size = 5000
msg=[]
```

```python
for i in wd:
  oh=one_hot(i,size)
  msg.append(oh)
```

```python
msg
```

```python
y = data['Is_Response']
```

```python
padding_len = 30
x = pad_sequences(msg,padding='post',maxlen=padding_len)
```

```python
x[0]
```

```python
from tensorflow.keras.layers import LSTM,Dense,Embedding
from tensorflow.keras import Sequential
```

```python

features = 200
```

```python
model = Sequential()
model.add(Embedding(size,features,input_length=padding_len))
model.add(LSTM(units=50,input_length=(30,50),activation='relu',return_sequences=True))
model.add(LSTM(units= 100,activation= 'relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='mean_absolute_error',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=5)
```

```python
predicts = ['I am very glad to see my mother happy again']
pred_data = []
for i in predicts:
  res = one_hot(i,100)
  pred_data.append(res)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_pred = pad_sequences(pred_data,maxlen = 30,padding='pre')
```

```python
result = model.predict(padded_pred)
```
